# ques2--> How does the overall quality of houses correlate with their sale prices?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a DataFrame named 'df'
# Load the dataset
df = pd.read_csv('/Users/tanishachopra/Downloads/house_sale_data.csv')

# Selecting columns for analysis
quality_saleprice = df[['OverallQual', 'SalePrice']]

# Calculate the correlation coefficient
correlation = quality_saleprice.corr().iloc[0, 1]

# Plotting the relationship between OverallQual and SalePrice
plt.figure(figsize=(10, 6))
plt.scatter(quality_saleprice['OverallQual'], quality_saleprice['SalePrice'], color='skyblue')
plt.title('Overall Quality vs Sale Price')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.grid(True)
plt.show()

print("Correlation coefficient between Overall Quality and Sale Price:", correlation)
```

# ques9--> What is the average living area (GrLivArea) for each house style (HouseStyle)?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset is stored in a CSV file named 'dataset.csv'
data = pd.read_csv('/Users/tanishachopra/Downloads/house_sale_data.csv')

# Filter the dataset to include only 'GrLivArea' and 'HouseStyle' columns
grlivarea_housestyle_data = data[['GrLivArea', 'HouseStyle']]

# Group the data by 'HouseStyle' and calculate the mean of 'GrLivArea'
average_grlivarea_by_housestyle = grlivarea_housestyle_data.groupby('HouseStyle')['GrLivArea'].mean().reset_index()

# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='HouseStyle', y='GrLivArea', data=average_grlivarea_by_housestyle, palette='viridis')
plt.title('Average Living Area by House Style')
plt.xlabel('House Style')
plt.ylabel('Average Living Area')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# ques10--> How does the presence of a pool affect the sale price?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming the dataset is stored in a CSV file named 'dataset.csv'
data = pd.read_csv('/Users/tanishachopra/Downloads/house_sale_data.csv')

# Filter the dataset to include only 'PoolArea' and 'SalePrice' columns
pool_saleprice_data = data[['PoolArea', 'SalePrice']]

# Create a new column indicating whether the property has a pool or not
pool_saleprice_data['HasPool'] = pool_saleprice_data['PoolArea'].apply(lambda x: 'Yes' if x > 0 else 'No')

# Create a pivot table to get the average sale price for each presence of a pool
pivot_table = pool_saleprice_data.pivot_table(values='SalePrice', index='HasPool', aggfunc='mean')

# Plot using Seaborn's heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt=".0f", cbar=True)
plt.title('Average Sale Price by Presence of Pool')
plt.xlabel('Presence of Pool')
plt.ylabel('Sale Price')
plt.show()
```

# ques13--> What is the average sale price per month (MoSold) or per year (YrSold)?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset is stored in a CSV file named 'dataset.csv'
data = pd.read_csv('/Users/tanishachopra/Downloads/house_sale_data.csv')

# Extract month and year from 'MoSold' and 'YrSold' columns
data['MoSold'] = pd.to_datetime(data['MoSold'], format='%m').dt.month_name()
data['YearSold'] = pd.to_datetime(data['YrSold'], format='%Y')

# Group the data by month or year and calculate the average sale price
average_price_per_month = data.groupby('MoSold')['SalePrice'].mean()
average_price_per_year = data.groupby('YearSold')['SalePrice'].mean()

# Plot using line plot
plt.figure(figsize=(12, 6))

# Plot average sale price per month
plt.subplot(2, 1, 1)
average_price_per_month.plot(kind='line', marker='o', color='blue')
plt.title('Average Sale Price per Month')
plt.xlabel('Month')
plt.ylabel('Average Sale Price')

# Plot average sale price per year
plt.subplot(2, 1, 2)
average_price_per_year.plot(kind='line', marker='o', color='green')
plt.title('Average Sale Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Sale Price')

plt.tight_layout()
plt.show()
```
